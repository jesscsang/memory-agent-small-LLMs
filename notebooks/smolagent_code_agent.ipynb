```python
from smolagents import LiteLLMModel, CodeAgent

model_id = "ollama_chat/llama3.2:1b-instruct-fp16"

# Does not matter what you set the api_key to as we are 
# using ollama with OpenAI's function
reasoning_model = LiteLLMModel(model_id, api_base="http://localhost:11434", num_ctx=8192)
reasoner = CodeAgent(tools=[], model=reasoning_model, add_base_tools=True)
```

```python
response = reasoner.run("What is weather in Calgary?", reset=False)
```
